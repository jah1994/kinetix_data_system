```python
# imports
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def ccd_noise_equation(t, n_star, be, de, rho, npix, pb,
                      D, gamma, h_obs, H=8000):
    '''
    t: exposure time (s)
    n_star: photons per t (e- . s^-1)
    be: background rate (e- . s^-1 . per pix)
    de: dark rate (e- . s^-1 . per pix)
    rho: readout noise (e-)
    npix: number of source aperture pixels
    pb: sky section pixels
    '''
    #X = 1.5
    signal = n_star * t
    shot_and_dark_var = t * (n_star + npix*(1 + (npix/pb))*(be + de))
    read_var = npix * (1 + (npix / pb)) * (rho**2)
    scint_var = 10e-5 * (D ** (-4/3)) * (1/t) * (np.cos(gamma) ** -3) * np.exp(-2 * h_obs / H)
    #scint_var = n_star * (0.09 * (X**1.5) * np.exp(-h_obs / H ) / (D**(2/3) * np.sqrt(2 * t)))**2
    #print(np.sqrt(shot_and_dark_var), signal * np.sqrt(scint_var))
    noise = np.sqrt(shot_and_dark_var + read_var + (signal**2 * scint_var))
    scint_ratio = np.sqrt((n_star * scint_var) / (shot_and_dark_var + read_var))
    #noise = np.sqrt(shot_and_dark_var + read_var + scint_var)
    #scint_ratio = np.sqrt((scint_var) / (shot_and_dark_var + read_var))
    return signal / noise, scint_ratio, [shot_and_dark_var, read_var, signal**2 * scint_var, signal]
```

```python
## computing n_star
# mVega = 0
#flam_Vega = 3.61e-11 # W m-2 nm-1 (V-band)
#lam = 550e-9 # nm (V-band)

flam_Vega = 2.177e-11 # W m-2 nm-1 (R-band)
mVega = 0.07 # (R-band)
lam = 644e-9 # nm (R-band)

def photon_rate(mag, mref, fref, lam, h=6.63e-34, c=3e8):
    
    '''
    mag: magnitude of target
    fref: monochromatic reference flux at lam
    lam: effective wavelength [m]
    h: planck constant
    c: speed of light
    '''

    # mref = 0 by definition
    #flux = fref * (10 ** (mag / (-2.5)))
    flux = fref * 10**(0.4 * (mref - mag))
    
    # number of photons in s-1 . m-2 . nm-1
    n_ = flux * lam / (h * c)
    
    return n_
    
def telescope_photon_rate(n_, D, del_lam=86, q=1):
    '''
    n_: photon rate
    D: telescope mirror diameter [m]
    del_lam: bandwidth [nm]
    q: overall efficiency (telscope/instrument/filter/detector)
    '''
    return n_ * np.pi * (D/2)**2 * del_lam * q
```

```python
def compute_theoretical_snr(mode, gamma, ap_diam, D_tel, h_obs):
    
    be = 0
    
    if mode == 'Sensitivity':
        de = 1.03 # speed mode = 3, sens = 1.03
        rho = 1.2 # speed mode = 2, sens = 1.2
        ts = 0.0125
    elif mode == 'Speed':
        de = 3 # speed mode = 3, sens = 1.03
        rho = 2 # speed mode = 2, sens = 1.2   
        ts = 0.003
    elif mode == 'TAOS':
        de = 50
        rho = 5
        ts = 0.05
        
        
    npix = ap_diam**2
    pb = ap_diam**2

    flam_Vega = 2.177e-11 # W m-2 nm-1 (R-band)
    lam = 644e-9 # nm (R-band)

    mags = np.linspace(9, 17, 100)
    snrs_theor = []#
    scint_ratio = []
    noise = []
    for mag in mags:
        #n_ = photon_rate(mag=mag, fref=flam_Vega, lam=658e-9, h=6.63e-34, c=3e8)
        n_ = photon_rate(mag=mag, mref=mVega, fref=flam_Vega, lam=lam, h=6.63e-34, c=3e8)
        #n_star = telescope_photon_rate(n_, D=D_tel, del_lam=400, q=0.7)  
        n_star = telescope_photon_rate(n_, D=D_tel, del_lam=200, q=1)  
        snrs_theor.append(ccd_noise_equation(ts, n_star, be, de, rho, npix, pb, D=D_tel, gamma=gamma, h_obs=h_obs)[0])
        scint_ratio.append(ccd_noise_equation(ts, n_star, be, de, rho, npix, pb, D=D_tel, gamma=gamma, h_obs=h_obs)[1])
        noise.append(ccd_noise_equation(ts, n_star, be, de, rho, npix, pb, D=D_tel, gamma=gamma, h_obs=h_obs)[2])
    return snrs_theor, mags, scint_ratio, noise
```

```python
#gamma0 = 60 * (np.pi / 180)
#gamma1 = 70 * (np.pi / 180)
gamma0 = 10 * (np.pi / 180)
gamma1 = 10 * (np.pi / 180)
ap_diam0 = 58
ap_diam1 = 30
snrs_theor0, mags, sr0, n0 = compute_theoretical_snr('Sensitivity', gamma0, ap_diam0, D_tel=2.1, h_obs=2e3)
snrs_theor1, mags, sr1, n1 = compute_theoretical_snr('Speed', gamma1, ap_diam1, D_tel=2.1, h_obs=2e3)
snrs_theor2, mags, sr2, n2 = compute_theoretical_snr('Sensitivity', gamma0, ap_diam0, D_tel=11.25, h_obs=4e3)
snrs_theor3, mags, sr3, n3 = compute_theoretical_snr('Speed', gamma1, ap_diam1, D_tel=11.25, h_obs=4e3)
snrs_theor4, mags, sr4, n4 = compute_theoretical_snr('TAOS', gamma0, ap_diam1, D_tel=1.3, h_obs=3e3)
```

```python
fs = 18
plt.figure(figsize=(15,10))
plt.plot(mags, snrs_theor0, c='k', linestyle='-',
         alpha=0.75, label='McDonald 2.1 m (Sensitivity mode: 80 Hz)')
plt.plot(mags, snrs_theor1, c='k', linestyle='--',
         alpha=0.75, label='McDonald 2.1 m (Speed mode: 300 Hz)')
plt.plot(mags, snrs_theor2, c='navy', linestyle='-', label='MSE (Sensitivity mode: 80 Hz)')
plt.plot(mags, snrs_theor3, c='navy', linestyle='--', label='MSE (Speed mode: 300 Hz)')
plt.plot(mags, snrs_theor4, c='darkorange', linestyle='-.', label='TAOS II (20 Hz)')
plt.grid()
plt.xlim(9, 17)
plt.tick_params(labelsize=fs)
plt.legend(fontsize='xx-large', loc='upper right')
plt.yscale('log')
plt.xlabel('r [magnitude]', fontsize=fs)
plt.ylabel('SNR', fontsize=fs);
#plt.savefig('SNR_curve.png', bbox_inches='tight')
```
