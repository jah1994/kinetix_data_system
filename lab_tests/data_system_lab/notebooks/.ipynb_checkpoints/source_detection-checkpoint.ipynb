```python
# imports
import numpy as np
import matplotlib.pyplot as plt
from astropy.stats import mad_std
from photutils.detection import find_peaks
%matplotlib inline
```

```python
N = 3200
ref = np.random.randn(N, N)
ref.shape
```

```python
plt.imshow(ref)
plt.colorbar();
```

```python
# compute estimates of the noise and sky level
sky = np.median(ref)
ref -= sky # sky subtract
std = mad_std(ref)
```

```python
# run initial peak finding routine on data to detect bright stars
peaks = find_peaks(ref, threshold=10*std, box_size=25, border_width=50)
```

```python
if peaks is None:
    print(peaks)
```
